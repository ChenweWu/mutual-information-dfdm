```python
import pandas as pd
import os
```

```python
    
df = pd.DataFrame()

for i, city in enumerate(os.listdir('cities/')):
    if city.endswith('.csv'):
        df_mun = pd.read_csv('cities/' + city)
        df_mun['Municipality code'] = city[:-4]
        df_mun = df_mun[['Municipality code', 'publish_date', 'title']]
        
        # concatenate all dataframes
        df = pd.concat([df, df_mun])
        
        
df
```

```python
df.to_csv('labels.csv', index=False)
```
