## Get gdelt dataset

```python
import pandas as pd
import os
```

```python
PATH_CITIES = '../../cities.csv'
PATH = 'Data/'
```

### Merge data

#### Get cities

```python
cities = pd.read_csv('../../cities.csv')
cities
```

#### Get datasets

```python
df = pd.DataFrame()
for city in os.listdir(PATH):
    if city.endswith('.csv'):
        city_path = os.path.join(PATH, city)
        df_aux = pd.read_csv(city_path)
        df_aux['Municipality code'] = city.replace('.csv', '')
        df = pd.concat([df, df_aux], ignore_index=True)
        
df
```

```python
df.to_csv('labels.csv', index=False)
```
